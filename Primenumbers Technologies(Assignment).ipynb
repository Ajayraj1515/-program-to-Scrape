```python
import requests
from bs4 import BeautifulSoup
import re

url = "https://qcpi.questcdn.com/cdn/posting/?group=1950787&provider=1950787"

response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")

search_postings_heading = soup.find(text=re.compile("Search Postings"))
if search_postings_heading is not None:
    search_postings_table = search_postings_heading.find_next_sibling("table")
    if search_postings_table is not None:
        rows = search_postings_table.find_all("tr")[1:]
        for i, row in enumerate(rows[:10]):
            columns = row.find_all("td")
            est_value = columns[0].text.strip()
            notes = columns[1].text.strip()
            description = columns[2].text.strip()
            closing_date = columns[3].text.strip()
            print(f"Posting {i+1} - Est. Value: {est_value}, Notes: {notes}, Description: {description}, Closing Date: {closing_date}")
else:
    print("Search Postings not found")
```
